```python
import requests, json
from datetime import datetime

link = requests.get("https://api.openweathermap.org/data/2.5/onecall?lat=-6.176274&lon=106.746540&exclude=hourly,minutely&units=metric&appid=ee1b11171a66052ac8fff6125177ef6b")
data = json.loads(link.text)
data
```

```python
#
soal 2
for i in range(len(data['daily'])):
    date = int(data['daily'][i]['dt'])
    object = datetime.utcfromtimestamp(date).strftime('%Y-%m-%d %H:%M:%S')
    datetimeobject = datetime.strptime(object,'%Y-%m-%d %H:%M:%S')
    print(datetimeobject.strftime('%a, %d %B %Y'), ": ", data['daily'][i]['temp']['day'],"°C")
```
